```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading dataset
file_path = './Resources/Top 50 Fast-Food Chains in USA.csv'
ffood = pd.read_csv(file_path)
ffood.head()
```

```python
#2023
file_path = './Resources/Top_50_Fast_Food(2024).csv'
ffood2024 = pd.read_csv(file_path)
ffood2024.head()
```

```python
# Renaming columns for easier access
ffood2024.columns = [
    "Rank",
    "Fast_Food_Chains",
    "Sales(Millions)",
    "Avg_Sales_Per_Unit(Thousands)",
    "Franchised_Stores",
    "Company_Stores",
    "Total_Units_2024",
    "Unit_Change_from_2024",
]
ffood2024.head()
```

```python
# Adjusting the formatting function
def format_chain_name(name):
    words = name.split()
    formatted_words = [words[0].capitalize()]  # Always capitalize the first word
    for word in words[1:]:
        if word.lower() == "'s":  # Ensure possessive "'s" is lowercase
            formatted_words.append("'s")
        else:
            formatted_words.append(word.capitalize())  # Capitalize other words
    return " ".join(formatted_words)

# Apply the corrected function
ffood2024['Fast_Food_Chains'] = ffood2024['Fast_Food_Chains'].apply(format_chain_name)

# Save the corrected dataset
ffood2024.to_csv("Resources/Top_50_Fast_Food_Cleaned(2024)_Corrected.csv", index=False)

# Confirm changes
print(ffood2024['Fast_Food_Chains'].head())
```

```python
import re

# Function to clean and standardize chain names further
def clean_chain_name(name):
    # Remove special characters and extra spaces
    name = re.sub(r'[’‘`]', "'", name)  # Replace curly quotes with straight quotes
    name = re.sub(r'&', 'and', name)  # Replace & with 'and'
    name = re.sub(r'[^\w\s\']', '', name)  # Remove other special characters
    name = re.sub(r'\s+', ' ', name).strip()  # Remove extra spaces
    return name.lower()

# Apply cleaning to both datasets
ffood['Fast_Food_Chains'] = ffood['Fast_Food_Chains'].apply(clean_chain_name)
ffood2024['Fast_Food_Chains'] = ffood2024['Fast_Food_Chains'].apply(clean_chain_name)

# Re-match datasets after cleaning
merged_ffood = pd.merge(ffood, ffood, on='Fast_Food_Chains', how='inner')

```

```python

# Merging the two datasets based on the company name
merged_ffood = pd.merge(
    ffood[["Fast_Food_Chains", "Total_Units_2021"]],
    ffood2024[["Fast_Food_Chains", "Total_Units_2024"]],
    on="Fast_Food_Chains",
    how="inner"
)

# Calculating the increase in units from 2021 to 2024
merged_ffood["Unit_Change"] = merged_ffood["Total_Units_2024"] - merged_ffood["Total_Units_2021"]
merged_ffood



```

```python
# Sorting the data by units increased for better visualization
merged_ffood_sorted = merged_ffood.sort_values(by="Unit_Change", ascending=False)

# Generate a gradient of colors (green to red) based on the index of the sorted dataframe
num_chains = len(merged_ffood_sorted)
colors = [
   (i / num_chains, 1 - i / num_chains, 0, 1) for i in range(num_chains)
]

# Plotting the chart with gradient colors
plt.figure(figsize=(12, 6))
plt.bar(merged_ffood_sorted["Fast_Food_Chains"], merged_ffood_sorted["Unit_Change"], color=colors, alpha=0.8)

# Adding labels and title
plt.xticks(rotation=90)
plt.xlabel("Fast-Food Chains")
plt.ylabel("Units Changed (2021 to 2024)")
plt.title("Change in Total Store Units(2021 to 2024)")

# Adjusting layout and displaying the plot
plt.tight_layout()
plt.savefig('./Resources/total_store_units(2021-2024)')
plt.show()
```
